# Dataset - All Weather By Hour

* `city_latitude` - Latitude coordinate of the city.
* `city_longitude` - Longitude coordinate of the city.
* `city_resolvedAddress` - Resolved address of the city, including locality, district, and country.
* `city_address` - Shortened address or name of the city.
* `city_timezone` - Timezone of the city.
* `city_tzoffset` - Timezone offset from UTC (Coordinated Universal Time) for the city.
* `day_datetime` - Date of the recorded data.
* `day_datetimeEpoch` - Epoch timestamp corresponding to day_datetime.
* `day_tempmax` - Maximum temperature recorded during the day.
* `day_tempmin` - Minimum temperature recorded during the day.
* `hour_visibility` - Visibility in kilometers at the given hour.
* `hour_cloudcover`- Percentage of sky covered by clouds at the given hour.
* `hour_solarradiation` - Solar radiation in Watts per square meter at the given hour.
* `hour_solarenergy` - Solar energy in Joules per square meter at the given hour.
* `hour_uvindex` - UV (Ultraviolet) index at the given hour.
* `hour_severerisk` - Risk level of severe weather at the given hour.
* `hour_conditions` - Weather conditions at the given hour.
* `hour_icon` - Icon representing weather conditions at the given hour.
* `hour_source` - Source of weather data for the given hour.
* `hour_stations` - Number of weather stations contributing data for the given hour.
* `etc` ...

# Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns                       
import matplotlib.pyplot as plt 
```

# Loading & Preprocessing Data

```python
# weather = pd.read_csv("all_weather_by_hour_v2.csv")
weather = pd.read_csv('..\clean_data\all_weather_by_hour_v2.csv')
weather
```

```python
weather.drop_duplicates(inplace=True)
```

```python
weather = weather.dropna()    
weather.count()
```

```python
for column, dtype in weather.dtypes.items():
    print(f"'{column}' - {dtype}")
```

```python
# city_resolvedAddress -> str
# city_address -> str
# city_timezone -> str
# day_datetime -> timestamp

columns_and_functions = {
    'city_resolvedAddress': str,
    'city_address': str,
    'city_timezone': str,
    'day_datetime': pd.to_datetime
}

for column, conversion_func in columns_and_functions.items():
    weather.loc[:, column] = weather[column].apply(conversion_func)
```

```python
for column, dtype in weather.dtypes.items():
    print(f"'{column}' - {dtype}")
```

# Visualization | EDA

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='day_conditions', x='day_temp', data=weather, hue='day_conditions', palette='coolwarm', legend=False)
plt.xlabel('Daily Temperature (°C)', fontsize=12, fontweight='bold')
plt.ylabel('Weather Conditions', fontsize=12, fontweight='bold')
plt.title('Daily Temperature Distribution by Weather Conditions', fontsize=14, fontweight='bold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mean_temp_by_condition = weather.groupby('day_conditions')['day_temp'].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_temp_by_condition.values, y=mean_temp_by_condition.index, hue=mean_temp_by_condition.index, palette='coolwarm', legend=False)
plt.xlabel('Mean Daily Temperature (°C)', fontsize=12, fontweight='bold')
plt.ylabel('Weather Conditions', fontsize=12, fontweight='bold')
plt.title('Mean Daily Temperature by Weather Conditions', fontsize=14, fontweight='bold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
seasons = {
    'Spring 2022': ('2022-03-01', '2022-05-31'),
    'Summer 2022': ('2022-06-01', '2022-08-31'),
    'Autumn 2022': ('2022-09-01', '2022-11-30'),
    'Winter 2022-2023': ('2022-12-01', '2023-02-28')
}
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
colors = ['linen', 'indianred', 'rosybrown', 'lavender']
for ax, (season, (start_date, end_date)), color in zip(axes.flatten(), seasons.items(), colors):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_data = weather[(weather['day_datetime'] >= start_date) & (weather['day_datetime'] <= end_date)]
    sns.histplot(season_data['day_humidity'], bins=20, kde=True, color=color, edgecolor='black', ax=ax)
    ax.set_title(f'Distribution of Humidity | {season}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Humidity (%)', fontweight='bold')
    ax.set_ylabel('Frequency', fontweight='bold')
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
colors = sns.color_palette("ch:s=.25,rot=-.25", len(seasons))
for i, (season, (start_date, end_date)) in enumerate(seasons.items()):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_data = weather[(weather['day_datetime'] >= start_date) & (weather['day_datetime'] <= end_date)]
    sns.lineplot(x=season_data['day_datetime'], y=season_data['day_humidity'], label=season, color=colors[i])
    mean_humidity = season_data['day_humidity'].mean()
    plt.axhline(mean_humidity, color=colors[i], linestyle='--', label=f'Mean Humidity ({season})')
plt.xlabel('Date', fontsize=12, fontweight='bold')
plt.ylabel('Humidity (%)', fontsize=12, fontweight='bold')
plt.title('Humidity Trend Over Time', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
colors = sns.color_palette("dark:#5A9_r", len(seasons))
for i, (season, (start_date, end_date)) in enumerate(seasons.items()):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_data = weather[(weather['day_datetime'] >= start_date) & (weather['day_datetime'] <= end_date)]
    sns.lineplot(x=season_data['day_datetime'], y=season_data['day_temp'], label=season, color=colors[i])
    mean_temp = season_data['day_temp'].mean()
    plt.axhline(mean_temp, color=colors[i], linestyle='--', label=f'Mean Temperature ({season})')
plt.xlabel('Date', fontsize=12, fontweight='bold')
plt.ylabel('Temperature (°C)', fontsize=12, fontweight='bold')
plt.title('Temperature Trend Over Time', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
colors = sns.color_palette("flare", len(seasons))
for i, (season, (start_date, end_date)) in enumerate(seasons.items()):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_data = weather[(weather['day_datetime'] >= start_date) & (weather['day_datetime'] <= end_date)]
    sns.lineplot(x=season_data['day_datetime'], y=season_data['hour_visibility'], label=season, color=colors[i])
    mean_visibility = season_data['hour_visibility'].mean()
    plt.axhline(mean_visibility, color=colors[i], linestyle='--', label=f'Mean Visibility ({season})')
plt.xlabel('Date', fontsize=12, fontweight='bold')
plt.ylabel('Visibility', fontsize=12, fontweight='bold')
plt.title('Visibility Trend by Season', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='day_temp', y='day_humidity', data=weather, color='royalblue', edgecolor='black')
plt.xlabel('Temperature (°C)', fontsize=12, fontweight='bold')
plt.ylabel('Humidity (%)', fontsize=12, fontweight='bold')
plt.title('Temperature vs. Humidity', fontsize=14, fontweight='bold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weather_group1 = weather[['day_temp', 'day_humidity', 'hour_visibility']]
correlation_matrix_group1 = weather_group1.corr()
weather_group2 = weather[['hour_solarradiation', 'hour_solarenergy', 'day_temp']]
correlation_matrix_group2 = weather_group2.corr()
weather_group3 = weather[['hour_uvindex', 'hour_cloudcover', 'day_humidity']]
correlation_matrix_group3 = weather_group3.corr()
```

```python
plt.figure(figsize=(23, 7))
plt.subplot(1, 3, 1)
sns.heatmap(correlation_matrix_group1, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Temperature, Humidity, Visibility', fontsize=14, fontweight='bold')
plt.subplot(1, 3, 2)
sns.heatmap(correlation_matrix_group2, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Solar Radiation, Solar Energy, Temperature', fontsize=14, fontweight='bold')
plt.subplot(1, 3, 3)
sns.heatmap(correlation_matrix_group3, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: UV Index, Cloud Cover, Humidity', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
