# Dataset - Alarms

* `id` — Unique identifier for each record
* `region_id` — Identifier for the region associated with the alarm
* `region_title` — Title or name of the region associated with the alarm
* `region_city` — City within the region associated with the alarm
* `all_region` — Indicates if the alarm is for the entire region (1 = yes, 0 = no)
* `start` — Start time of the alarm
* `end` —  End time of the alarm
* `clean_end` — Cleaned end time of the alarm (if applicable)
* `intersection_alarm_id` — Identifier for intersection alarm (if applicable)

# Visualization

```python
sns.set(font='DejaVu Sans')
alarms_per_region = alarms['region_title'].value_counts()
alarms_per_region = alarms_per_region.sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=alarms_per_region.values, y=alarms_per_region.index, palette='viridis')
for i, v in enumerate(alarms_per_region.values):
    plt.text(v + 3, i, f'{v}', va='center')
plt.axvline(x=alarms_per_region.mean(), color='red', linestyle='--', label='Mean Value')
plt.title('Number of Alarms per Region', fontsize=14, fontweight='bold')
plt.xlabel('Number of Alarms', fontsize=12, fontweight='bold')
plt.ylabel('Region', fontsize=12)
plt.grid(axis='x', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)
plt.legend()
plt.show()
```

```python
seasons = {'Spring 2022': ('2022-03-01', '2022-05-31'),
           'Summer 2022': ('2022-06-01', '2022-08-31'),
           'Autumn 2022': ('2022-09-01', '2022-11-30'),
           'Winter 2022-2023': ('2022-12-01', '2023-02-28')}
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
colors = ['linen', 'indianred', 'rosybrown', 'lavender']
for ax, (season, (start_date, end_date)), color in zip(axes.flatten(), seasons.items(), colors):
    season_alarms = alarms[(alarms['start'] >= start_date) & (alarms['end'] <= end_date)]
    alarm_counts = season_alarms.groupby('region_title').size().sort_values(ascending=False)
    regions = alarm_counts.index
    count = alarm_counts.values
    ax.barh(regions, count, color=color, edgecolor='black')
    ax.set_title(f'Number of Alarms by Region | {season}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Count', fontweight='bold')
    ax.set_ylabel('Region', fontweight='bold')
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
alarms['duration'] = (alarms['end'] - alarms['start']).dt.total_seconds() / 60
average_duration_per_region = alarms.groupby('region_title')['duration'].mean().reset_index()
fig, axs = plt.subplots(1, 2, figsize=(15, 8))
sns.scatterplot(data=average_duration_per_region, x='duration', y='region_title', hue='duration', palette='coolwarm', s=100, ax=axs[0])
axs[0].set_xlabel('Average Alarm Duration (minutes)', fontsize=12, fontweight='bold')
axs[0].set_ylabel('Region', fontsize=12, fontweight='bold')
axs[0].set_title('Average Alarm Duration by Region', fontsize=14, fontweight='bold')
axs[0].legend().remove()
sns.boxplot(x='duration', y='region_title', data=alarms, palette='coolwarm', ax=axs[1])
axs[1].set_xlabel('Alarm Duration (minutes)', fontsize=12, fontweight='bold')
axs[1].set_ylabel('Region', fontsize=12, fontweight='bold')
axs[1].set_title('Alarm Duration Distribution by Region', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
season_weekday_counts = {season: {weekday: 0 for weekday in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']} for season in seasons}
for season, (start_date, end_date) in seasons.items():
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_alarms = alarms[(alarms['start'] >= start_date) & (alarms['start'] <= end_date)]
    weekday_counts = season_alarms['start'].dt.day_name().value_counts()
    for weekday, count in weekday_counts.items():
        season_weekday_counts[season][weekday] = count
plt.figure(figsize=(18, 8))
palette = sns.color_palette("seismic", len(seasons))
for i, (season, counts) in enumerate(season_weekday_counts.items()):
    plt.plot(list(counts.keys()), list(counts.values()), marker='o', label=season, color=palette[i])
plt.title('Distribution of Alarms Over Weekdays for Different Seasons', fontsize=14, fontweight='bold')
plt.xlabel('Weekday', fontweight='bold')
plt.ylabel('Number of Alarms', fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
season_hour_counts = {season: {hour: 0 for hour in range(24)} for season in seasons}
for season, (start_date, end_date) in seasons.items():
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    season_alarms = alarms[(alarms['start'] >= start_date) & (alarms['start'] <= end_date)]
    hour_counts = season_alarms['start'].dt.hour.value_counts().sort_index()
    for hour, count in hour_counts.items():
        season_hour_counts[season][hour] = count
palette = sns.color_palette('seismic', n_colors=len(seasons))
plt.figure(figsize=(18, 8))
for i, (season, counts) in enumerate(season_hour_counts.items()):
    plt.plot(list(counts.keys()), list(counts.values()), marker='o', label=season, color=palette[i])
plt.title('Distribution of Alarms Over Hours of the Day for Different Seasons', fontsize=14, fontweight='bold')
plt.xlabel('Hour of the Day', fontweight='bold')
plt.ylabel('Number of Alarms', fontweight='bold')
plt.legend()
plt.grid(True)
plt.xticks(range(24))
plt.show()
```
